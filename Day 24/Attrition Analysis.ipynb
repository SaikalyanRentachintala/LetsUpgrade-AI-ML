# Attrition Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
data=pd.read_csv("general_data.csv")
data.head()
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data["NumCompaniesWorked"]=data["NumCompaniesWorked"].fillna(data["NumCompaniesWorked"].mean())
data["TotalWorkingYears"]=data["TotalWorkingYears"].fillna(data["TotalWorkingYears"].mean())
```

```python
data.isna().sum()
```

```python
data.hist(figsize=(10,10))
plt.show()
```

# Inference
Employee Count,Standard Hours,Employee Id are not useful for the Analysis.

Education, Job Level, Stock Option Level are Categorical Values.

```python
col_list=['Age', 'DistanceFromHome', 'MonthlyIncome', 'NumCompaniesWorked', 'PercentSalaryHike','TotalWorkingYears', 
           'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
for col in col_list:
    sb.distplot(data[col])
    plt.show()
```

# Inference
Age, NumCompaniesWorked, PercentSalaryHike, TotalWorkingYears, YearsSinceLastPromotion is Unimodal.

DistanceFromHome is Bimodal.

MonthlyIncome, TrainingtimesLastYear, YearsAtCompany, YearswithCurrManager is Multimodal.

```python
column_list = ['Age', 'DistanceFromHome', 'Education', 'EmployeeCount', 'EmployeeID','JobLevel','MonthlyIncome', 
       'NumCompaniesWorked', 'PercentSalaryHike', 'StandardHours','StockOptionLevel',
       'TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager']

for a in column_list:
    sb.boxplot(data[a])
    plt.show()
```

# Inference
No Outliers: Age, DistanceFromHome, Education, EmployeeID, JobLevel, PercentSalaryHike.

Outliers: TrainingTimesLastYear, YearsWithCurrManager, MonthlyIncome, NumCompaniesWorked, StockOptionLevel.

Irrelevant Data: EmployeeCount , StandardHours.

# Converting Object type Columns to Int

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

data['Attrition'] = le.fit_transform(data['Attrition'])
data['BusinessTravel'] = le.fit_transform(data['BusinessTravel'])
data['Department'] = le.fit_transform(data['Department'])
data['Over18'] = le.fit_transform(data['Over18'])
data['MaritalStatus'] = le.fit_transform(data['MaritalStatus'])
data['JobRole'] = le.fit_transform(data['JobRole'])
data['Gender'] = le.fit_transform(data['Gender'])
data['EducationField'] = le.fit_transform(data['EducationField'])
```

```python
data = data.drop(['EmployeeCount','EmployeeID', 'Over18','StandardHours'], axis=1)
data.head()
```

```python
y_log=data.iloc[:,1:2]
y_log.head()
```

```python
x_log = data.drop(['Attrition'], axis=1)
x_log.head()
```

# Decision Tree

```python
from sklearn import tree
from sklearn import preprocessing
```

```python
model = tree.DecisionTreeClassifier()
model.fit(x_log,y_log)
```

```python
model.score(x_log,y_log)
```

__Accuracy is 100% possibility that data is overfitted.__

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=1000, max_features=2, oob_score=True)
```

```python
rf.fit(x_log,y_log)
```

```python
rf.oob_score_
```

# Finding Important Columns for predicting Attrition

```python
a = ['Age', 'BusinessTravel', 'Department', 'DistanceFromHome','Education', 'EducationField', 'Gender', 'JobLevel', 'JobRole',
       'MaritalStatus', 'MonthlyIncome', 'NumCompaniesWorked','PercentSalaryHike', 'StockOptionLevel', 'TotalWorkingYears',
       'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion','YearsWithCurrManager']
for feature,imp in zip(a, rf.feature_importances_):
    print(feature,"\t\t :",imp);
```

__Inference:__

Age, DistanceFromHome, MonthlyIncome, PercentSalaryHike, TotalWorkingYears, YearsAtCompany value is more than 0.05 so, they are the most important columns required.

```python
imp_col = data[['Age','DistanceFromHome','MonthlyIncome','PercentSalaryHike','TotalWorkingYears','YearsAtCompany']]
imp_col.head()
```

```python
model = tree.DecisionTreeClassifier(max_depth=12)
model.fit(imp_col,y_log)
```

# Creating Tree

```python
from sklearn.tree import export_graphviz
from IPython.display import SVG
from graphviz import Source
from IPython.display import display
```

```python
graph = Source(tree.export_graphviz(model, out_file=None, 
                                    feature_names=['Age','DistanceFromHome','MonthlyIncome','PercentSalaryHike',
                                                   'TotalWorkingYears','YearsAtCompany'],
                                    class_names=['0','1'] , filled = True))
display(SVG(graph.pipe(format='svg')))
```

# Predicting Score

```python
model.score(imp_col,y_log)
```

# Fitting Random Forest

```python
rf.fit(imp_col,y_log)
```

# Accuracy Score

```python
rf.oob_score_
```

# Inference

Accuracy Score is same in case of both Decision Tree and Random Forest so, it means data is not over fitted in the case of Decision Tree.
